```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np

def adaptive_edge_detection(image_path, threshold1, threshold2):
    # Load the image
    img = cv2.imread(image_path, 0)  # Read image as grayscale

    # Apply Gaussian blur to reduce noise
    img_blur = cv2.GaussianBlur(img, (3, 3), 0)

    # Adaptive thresholding to detect edges
    edges = cv2.Canny(img_blur, threshold1, threshold2)

    # Display the original image and edges
    cv2_imshow(img)
    cv2_imshow(edges)

# Example usage
image_path = '/content/IMG_20180413_083129-01.jpeg'
threshold1 = 50  # Lower threshold for edge detection
threshold2 = 150  # Upper threshold for edge detection

adaptive_edge_detection(image_path, threshold1, threshold2)
```
